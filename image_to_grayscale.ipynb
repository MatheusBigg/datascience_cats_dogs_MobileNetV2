```python
from PIL import Image
from google.colab import files
```

```python
#upar imagem

uploaded = files.upload()
caminho_entrada = "gato_tranquilo.jpg"
imagem = Image.open(caminho_entrada).convert("RGB")
```

```python
imagem
```

```python
#Converter pra cinza

largura, altura = imagem.size
pixels = imagem.load()

imagem_cinza = Image.new("RGB", (largura, altura))

for y in range(altura):
    for x in range(largura):
        r, g, b = pixels[x, y]
        # Frmula ponderada
        gray = int(0.299 * r + 0.587 * g + 0.114 * b)
        imagem_cinza.putpixel((x, y), (gray, gray, gray))

imagem_cinza
```

```python
#Converter pra PB

threshold = 128
largura, altura = imagem_cinza.size
pixels = imagem_cinza.load()

imagem_pb = Image.new("RGB", (largura, altura))

for y in range(altura):
    for x in range(largura):
        r, g, b = pixels[x, y]
        # Todos canais so iguais no cinza, usar r
        bw = 255 if r >= threshold else 0
        imagem_pb.putpixel((x, y), (bw, bw, bw))

imagem_pb
```
